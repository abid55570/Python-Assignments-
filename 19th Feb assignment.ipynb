Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a Python web framework that is designed to create web applications quickly and easily. It is a micro-framework, which means that it is small and lightweight and does not include everything that a larger framework like Django might have. However, Flask is highly customizable and extensible, allowing developers to add in the specific features they need for their application.

Advantages of using Flask Framework:

Lightweight: Flask is a micro-framework that doesn't require any particular tools or libraries to run. It is small and simple, making it easy to learn and use.

Flexible: Flask allows developers to add and customize features according to their specific needs. Developers can choose which libraries and tools they want to use, making it highly adaptable to any project.

Easy to learn: Flask has a minimalistic design, making it easy for developers to learn and use. It requires less boilerplate code and is easy to set up and get started with.

Extensible: Flask is highly extensible and allows developers to add plugins, extensions, and other third-party tools to enhance its functionality.

Great community support: Flask has a large and active community of developers who contribute to its development, offer support, and create new tools and libraries that can be easily integrated into Flask applications.

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing is the process of mapping URLs or routes to specific functions in your application. This allows you to create different pages or endpoints within your web application and define what happens when a user navigates to those pages.

To define a route in Flask, you use the @app.route decorator, followed by the URL or route you want to map to the function. For example:

@app.route('/')
def home():
    return 'This is the home page!'

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

The url_for() function is used in Flask to build URLs for specific routes in your application. It takes the name of the function as its first argument and any additional arguments as keyword arguments.

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page!'

@app.route('/about')
def about():
    return 'This is the about page!'

@app.route('/contact')
def contact():
    return 'This is the contact page!'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))
        print(url_for('about'))
        print(url_for('contact'))
