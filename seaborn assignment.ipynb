Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
Seaborn is a popular Python library for statistical data visualization. It provides a high-level interface for creating informative and visually appealing plots. Here are five commonly used plots in Seaborn and their uses:

Scatter Plot: A scatter plot represents the relationship between two numeric variables. It uses dots to visualize the distribution of data points and the correlation between variables.

Uses:

Identifying patterns or trends between variables.
Examining the strength and direction of the relationship between variables.
Bar Plot: A bar plot displays categorical data using rectangular bars. The length of each bar corresponds to the value of the category it represents. Bar plots are useful for comparing values across different categories.

Uses:

Comparing the magnitude of different categories.
Visualizing the frequency or count of categorical variables.
Histogram: A histogram is a graphical representation of the distribution of a continuous variable. It uses bars to display the frequency or count of values falling into different bins or intervals.

Uses:

Understanding the shape, center, and spread of a continuous variable.
Identifying the presence of outliers or skewed distributions.
Heatmap: A heatmap is a matrix-like plot that visualizes the data using colors. It is particularly useful for displaying the correlation between variables in a tabular form.

Uses:

Visualizing patterns and relationships in a matrix-like dataset.
Identifying the strength and direction of correlations between variables.
Violin Plot: A violin plot is a combination of a box plot and a kernel density plot. It displays the distribution of a numeric variable across different categories as a vertical or horizontal violin-shaped plot.

Uses:

Visualizing the distribution of data and its summary statistics.
Comparing the distribution of a variable across different categories.
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_dataset = sns.load_dataset("fmri")

# Plot a line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_dataset)

# Set the title and labels
plt.title("Signal over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
# Load the "titanic" dataset
titanic_dataset = sns.load_dataset("titanic")

# Plot box plots
plt.figure(figsize=(10, 5))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_dataset)
plt.title("Box Plot of Age by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Age")

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_dataset)
plt.title("Box Plot of Fare by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Fare")

# Adjust the layout and display the plots
plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
# Load the "diamonds" dataset
diamonds_dataset = sns.load_dataset("diamonds")

# Plot histogram with hue
sns.histplot(data=diamonds_dataset, x="price", hue="cut")

# Set the title and labels
plt.title("Histogram of Price by Cut")
plt.xlabel("Price")
plt.ylabel("Count")

# Show the plot
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns

# Load the "iris" dataset
iris_dataset = sns.load_dataset("iris")

# Plot pair plot with hue
sns.pairplot(data=iris_dataset, hue="species")

# Show the plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Load the "flights" dataset
flights_dataset = sns.load_dataset("flights")

# Reshape the data for heatmap
flights_pivot = flights_dataset.pivot("month", "year", "passengers")

# Plot heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu")

# Set the title and labels
plt.title("Passenger Flights")
plt.xlabel("Year")
plt.ylabel("Month")

# Show the plot
plt.show()
```
